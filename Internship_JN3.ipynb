```python
import pandas as pd
import numpy as np
import scipy
```

```python
from scipy import signal
```

```python
dataset=pd.read_excel('Bern barcelona focal EEGs subset1.xlsx')
```

```python
import matplotlib.pyplot as plt
```

```python
resampled_dataset = scipy.signal.resample(dataset,3472)
```

```python
plt.figure(figsize=(12,7))
plt.plot(dataset,'yellow')
plt.plot(resampled_dataset,'red')
plt.legend(['dataset','resampled_data'],loc='best')
plt.xlabel('No. of samples')
plt.ylabel('Voltage potentials in mV') 
plt.title('Bern barcelona resampled data and orginal data')
plt.show()
```
