```python
rows % 22
```

```python
arr=[]
eigenvalues=[]
k=0
while(k<rows):
    for i in range(10):
        arr.append(list(data1[i+k]))
    eigen=np.linalg.eigvals(arr[k:k+22]) 
    eigenvalues.append(eigen)
    k=k+22
print(arr)
```

```python
print(eigenvalues)
```

```python
np.savetxt("EigenVector.cvs",eigenvalues,delimiter=',',fmt=('%.18d'))
```
