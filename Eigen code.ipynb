```python
%matplotlib qt5
import numpy as np
from numpy import zeros, arange
import csv
import pandas as pd
```

```python
df1=pd.read_csv('Book3.csv',header=None)
```

```python
data1=np.array(df1)
```

```python
data1.shape
```

```python
rows,cols=data1.shape
```

```python
rows % 22
```

```python
arr=[]
eigenvalues=[]
k=0
while(k<rows-20):
    for i in range(22):
        arr.append(list(data1[i+k]))
    eigen=np.linalg.eigvals(arr[k:k+22]) 
    eigenvalues.append(eigen)
    k=k+22
```

```python
print(eigenvalues)
```

```python
np.savetxt("C:/Users/abc/Desktop/Research paper/eigen_sheet.csv",eigenvalues,delimiter=',',fmt=('%.18d'))
```
